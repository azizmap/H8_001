### merging dataframe

#### concat

```python
city_data = pd.DataFrame(data=[{"revenue":1000, 'employee_count':200},
                                {"revenue":2000, 'employee_count':300},
                                {"revenue":3000, 'employee_count':400},
                               ], index=['Amsterdam','Tokyo', 'Toronto'])
city_dat2 = pd.DataFrame(data=[{"revenue":4000, 'employee_count':200},
                                {"revenue":5000, 'employee_count':300},
                               ], index=['Paris', 'Jakarta'])
```

```python
pd.concat([city_data, city_dat2])
```

```python
city_area = pd.DataFrame(data=[100,150,200],index=['Amsterdam','Tokyo', 'Toronto'])
```

```python
pd.concat([city_data, city_area], axis=1)
```

#### merge

```python
employees = pd.DataFrame(data=[{'employee_id':1,'name':'Budi'},{'employee_id':2,'name':'Sisi'}])
```

```python
employees_gender = pd.DataFrame(data=[{'employee_id':1,'gender':'M'},{'employee_id':2,'gender': 'L'}])
```

```python
pd.merge(employees,employees_gender,on='employee_id',how='inner')
```

## Exercise

```python
#visualize top 10 team most win in home game
top10_home = df[df['game_location'] == 'H']
top10_home = top10_home[top10_home['game_result'] == 'W']

top10_home = top10_home['team_id'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Team with Most Win (Home)')
plt.show()
```

```python
top10_home = df[df['game_location'] == 'H']
top10_home = top10_home[top10_home['game_result'] == 'W']
top10_home = top10_home.groupby(['team_id'])['team_id']
top10_home
```

```python
#visualize Lakers(LAL) number wins from 2010
lal = df[df['team_id'] == 'LAL']
lal = lal[lal['game_result'] == 'W']
lal = lal[lal['year_id'] >= 2010]

lal['year_id'].value_counts().head(10).plot()
plt.title('Lakers(LAL) Number of Wins since 2010')
plt.show()
```
