### Matplotlib

```python
matplotlib.style.use('ggplot')
```

```python
years = [str(x) for x in range(1980,2013+1)]
```

```python
%matplotlib inline
```

```python
df_can.loc['Haiti', years].plot(kind='line', figsize=(12,4))

plt.title('Immigrations from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.text(25,6000, '2010 Earthquake')

plt.show()
```

```python
China_India = df_can.loc[['China', 'India'], years]
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()

df_CI.plot(kind='line', figsize=(8,6))
plt.title('Immigrants from China and India')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

###### Top 5 Countries that contribute the most to the number of immigration to Canada

```python
df_can.sort_values('Total', ascending=False)[:5][years].transpose().plot(kind='line', figsize=(16,8))
plt.title('Top 5 Countries that contribute the most to the number of immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

###### Area Plot

```python
df_can.sort_values('Total', ascending=False)[:5][years].transpose().plot(kind='area', figsize=(16,8), alpha=0.5, stacked=False)
plt.title('Top 5 Countries that contribute the most to the number of immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_can.sort_values('Total', ascending=False)[:5][years].transpose().plot(kind='area', figsize=(16,8), alpha=0.5, stacked=True)
plt.title('Top 5 Countries that contribute the most to the number of immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
ax = df_can.sort_values('Total', ascending=False)[:5][years].transpose().plot(kind='area', figsize=(16,8), alpha=0.5, stacked=True)
ax.set_title('Top 5 Countries that contribute the most to the number of immigration to Canada')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')

plt.show()
```

```python
ax = df_can.sort_values('Total', ascending=False)[-5:][years].transpose().plot(kind='area', figsize=(16,6), alpha=0.5, stacked=False)
ax.set_title('The Least 5 Countries that contribute to the number of immigration to Canada')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')

plt.show()
```

###### Histogram

```python
plt.hist(df_can['2013'])
plt.show()
```

```python
df_can['2013'].plot(kind='hist', bins=300, figsize=(16,8))
plt.title('Histogram of Immigration from 195 Country in 2013')
plt.xlabel('Number of Immigrant')
plt.ylabel('Number of Country')
plt.show()
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10,6), alpha=0.4, bins=15, color=['coral','darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, Sweden from 1980-2013')
plt.xlabel('Number of Immigrant')
plt.ylabel('Number of Country')

plt.show()
```

###### Bar Chart

```python
df_Iceland = df_can.loc['Iceland', years]
```

```python
df_Iceland.plot(kind='bar', figsize=(10,6))

plt.title('Icelandic Immigrants to Canada 1980-2013')
plt.xlabel('Number of Immigrant')
plt.ylabel('Number of Country')

plt.show()
```

```python
df_can.sort_values(by='Total', ascending=True, inplace=True)
df_top15 = df_can['Total'].tail(15)
```

```python
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(10,6), color='blue')

plt.title('Top 5 Countries that contribute the most to the number of immigration to Canada')
plt.xlabel('Number of Immigrant')
plt.ylabel('Country')

plt.show()
```

###### Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(12,12),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True)

plt.title('Immigration to Canada by Continent 1980-2013')
plt.show()
```

###### Boxplot

```python
df_japan = df_can.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
df_japan.plot(kind='box', figsize=(8,6))

plt.title('Boxplot of Japanese Immigrants 1980-2013')
plt.ylabel('Number of Immigrant')

plt.show()
```

```python
df_CI.plot(kind='box', figsize=(8,6))

plt.title('Boxplot of China and India Immigrants 1980-2013')
plt.ylabel('Number of Immigrant')

plt.show()
```

```python
df_CI.plot(kind='box', figsize=(8,6), vert=False)

plt.title('Boxplot of China and India Immigrants 1980-2013')
plt.ylabel('Number of Immigrant')

plt.show()
```

###### Subplot

```python
fig = plt.figure(figsize=(20,8))

plt.suptitle('Super Title')

ax0 = fig.add_subplot(121)
ax1 = fig.add_subplot(122)

df_Iceland.plot(kind='bar', ax=ax0)
ax0.set_title('Icelandic Immigrants to Canada 1980-2013')
ax0.set_xlabel('Number of Immigrant')
ax0.set_ylabel('Number of Country')

df_CI.plot(kind='box', vert=False, ax=ax1)
ax1.set_title('Boxplot of China and India Immigrants 1980-2013')
ax1.set_ylabel('Number of Immigrant')

plt.show()
```

###### Scatter Plot

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.index = map(int, df_tot.index)

df_tot.reset_index(inplace=True)

df_tot.columns = ['year', 'total']

df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10,6))

plt.show()
```
